```python
medicine=pd.read_csv('https://raw.githubusercontent.com/chop-analytics/analyst-take-home-task/master/datasets/medications.csv',parse_dates=['START', 'STOP'])
```

```python
medicine=medicine[medicine.PATIENT.isin(encounters_overdose_final.PATIENT_ID)]
#encounters_overdose_final.PATIENT_ID
```

```python
medicine.shape
```

```python
medicine=medicine.loc[:,['START','STOP','PATIENT','DESCRIPTION']]
```

```python
medict = medicine.groupby('PATIENT')[
    'START', 'STOP', 'DESCRIPTION'].apply(
        lambda x: x.values.tolist()).to_dict()
```

```python
medict
```

```python
type(medict)
```

```python
medict
```

```python
encounters_overdose_final
```

```python
drug_encounter_dict={}

opioid_dict={}
op_list=['Hydromorphone 325 MG','Oxycodone-acetaminophen 100ML','Fentanyl 100 MCG']

for row in encounters_overdose_final.itertuples():
    if row.PATIENT_ID in medict:
        
        # y[0] start timestamp, y[1] stop timestamp y[2] description for medication
        # HOSPITAL_ENCOUNTER_DATE 
        for y in medict[row.PATIENT_ID]:
            
            if y[0] < row.HOSPITAL_ENCOUNTER_DATE :
                
                if row.Index in drug_encounter_dict:
                    drug_encounter_dict[row.Index] += 1
                else:
                    drug_encounter_dict[row.Index] = 1
                
                if y[2] in op_list:
                    opioid_dict[row.Index] = 1
            
```

```python
drug_encounter_dict
```

```python
opioid_dict
```

```python
encounters_overdose_final['COUNT_CURRENT_MEDS'] = encounters_overdose_final.index.map(drug_encounter_dict)
```

```python
encounters_overdose_final['CURRENT_OPIOID_IND'] = encounters_overdose_final.index.map(opioid_dict)
```

```python
encounters_overdose_final['CURRENT_OPIOID_IND'].fillna(0,inplace=True)
```

```python
encounters_overdose_final
```

```python
encounters_overdose_final.reset_index(inplace=True)
encounter_by_patient = encounters_overdose_final.groupby('PATIENT_ID')['ENCOUNTER_ID'].apply(lambda x: x.values)
```

```python
encounter_by_patient
```

```python
encounters_overdose_final.set_index('ENCOUNTER_ID',inplace=True)
```

```python
readmission_dates={}
readmin_30d={}
for i in encounter_by_patient:
    if len(i) > 1 :
        for j in range (len(i)-1): #for each encounter and compare with the next encounter
            time_diff=encounters_overdose_final.HOSPITAL_ENCOUNTER_DATE[i[j+1]] - encounters_overdose_final.STOP[i[j]]
            if time_diff < np.timedelta64(91,'D'):
                readmission_dates[i[j]] = encounters_overdose_final.HOSPITAL_ENCOUNTER_DATE[i[j+1]]
                if time_diff < np.timedelta64(31,'D'):
                    readmin_30d[i[j]]=1
                
            
```

```python
readmin_30d
```

```python
encounters_overdose_final['FIRST_READMISSION_DATE'] = encounters_overdose_final.index.map(readmission_dates)
```

```python
readmin_90 = ~pd.isnull(encounters_overdose_final['FIRST_READMISSION_DATE'].values)
```

```python
encounters_overdose_final.loc[readmin_90,'READMISSION_90_DAY_IND']=1
```

```python
encounters_overdose_final['READMISSION_30_DAY_IND'] = encounters_overdose_final.index.map(readmin_30d)
```

```python
encounters_overdose_final.reset_index(inplace=True)
```

```python
encounters_overdose_final
```

```python
# Drop stop column
encounters_overdose_final.drop('STOP', axis=1, inplace=True)

# change column order
encounters_overdose_final = encounters_overdose_final[(['PATIENT_ID',
                          'ENCOUNTER_ID',
                          'HOSPITAL_ENCOUNTER_DATE',
                          'AGE_AT_VISIT',
                          'DEATH_AT_VISIT_IND',
                          'COUNT_CURRENT_MEDS',
                          'CURRENT_OPIOID_IND',
                          'READMISSION_90_DAY_IND',
                          'READMISSION_30_DAY_IND',
                          'FIRST_READMISSION_DATE'])]
```

```python
encounters_overdose_final['COUNT_CURRENT_MEDS'].fillna(0,inplace=True)
#encounters_overdose_final.COUNT_CURRENT_MEDS
```

```python
 encounters_overdose_final['DEATH_AT_VISIT_IND'].fillna(0,inplace=True)
```

```python
#encounters_overdose_final.READMISSION_90_DAY_IND
encounters_overdose_final['READMISSION_90_DAY_IND'].fillna(0,inplace=True)
```

```python
encounters_overdose_final['READMISSION_30_DAY_IND'].fillna(0,inplace=True)
```

```python
encounters_overdose_final['FIRST_READMISSION_DATE'].fillna('N/A',inplace=True)
```
